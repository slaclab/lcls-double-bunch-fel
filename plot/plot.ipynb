```python
import matplotlib.pyplot
import numpy

dac_res = 16
max_dac = 2 ** dac_res - 1
half_dac = max_dac / 2.0
data_type = numpy.uint16

amp = 1    
segLen = 4096
cycles = 100
x = numpy.linspace(0, segLen-1, segLen)
omega = 2 * numpy.pi * cycles
y = (numpy.sin(omega*x/segLen) + 1.0) * half_dac
y = numpy.round(y)
y = numpy.clip(y, 0, max_dac)
y = y.astype(data_type)

matplotlib.pyplot.plot(x, y)
matplotlib.pyplot.show()
```
