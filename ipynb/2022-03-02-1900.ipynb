```python
# Figure out controls for changing pulse width

import numpy
import matplotlib.pyplot as pyplot
from matplotlib import pyplot

from ipywidgets import *


def get_pulse(width):
    linear = -0.03
    cubic = 0.0005
    offset = 10
    amplitude = 1

    length = 4096
    start = -10
    end = start + length

    # We need x values in units of nanoseconds to offset
    x = numpy.arange(start, end) / 1.428 
    x = x - offset

    x = x * (30 / (abs(width) + 1))
    w = ( (- numpy.tanh(x - 5) - numpy.tanh(-x - 12)) *
         (1 + linear*x + cubic*x**3) )

    w = w * amplitude / 2
    
    return w

figure, ax = pyplot.subplots(dpi=100)
ax.set_xlim(0, 40)

w = get_pulse(5)

ax.plot(w)

pyplot.show()
```
